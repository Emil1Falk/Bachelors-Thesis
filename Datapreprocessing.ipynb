```python
import pandas as pd
import string
import unicodedata
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/Users/emilfalk/Desktop/nba_stats_2020_21.csv', delimiter = ';') 
```

```python
df['Team'] = pd.factorize(df['Team'])[0]
```

```python
def clean_text(text):
    text = ''.join(
        c for c in unicodedata.normalize('NFD', text)
        if unicodedata.category(c) != 'Mn'
    )
    text = text.replace('Jr.', 'Jr').replace('.', '')
    if ' III' in text:
        text = text.replace(' III', '')
    return text
```

```python
df['Player'] = df['Player'].apply(clean_text)
```

```python
url = "https://hoopshype.com/salaries/players/2021-2022/"
tables = pd.read_html(url)
salary = tables[0]
salary = salary.rename(columns={'2021/22': 'Salary'})
salary = salary[['Player', 'Salary']]
salary['Salary'] = salary['Salary'].str.replace("$", "", regex=False).str.replace(",", "", regex=False).astype(int)
```

```python
df = pd.merge(df, salary, left_on='Player', right_on='Player', how='inner')
```

```python
df['ig_followers'] = df['ig_followers'].astype(str).str.replace('\xa0', '').str.replace(' ', '')
df['ig_followers'] = pd.to_numeric(df['ig_followers'], errors='coerce')
```

```python
X = df.drop(columns=['Salary'])  
y = df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train_df = pd.DataFrame(X_train)
X_test_df = pd.DataFrame(X_test)
y_train_df = pd.Series(y_train, name='Salary')
y_test_df = pd.Series(y_test, name='Salary')
```

```python
X_train_df.to_csv('X_train.csv', index=False)
X_test_df.to_csv('X_test.csv', index=False)
y_train_df.to_csv('y_train.csv', index=False)
y_test_df.to_csv('y_test.csv', index=False)
```
